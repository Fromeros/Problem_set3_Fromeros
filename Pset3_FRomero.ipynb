# Problem 1 (5 pts):

Write a script (or Jupyter Notebook code block) that opens the file, uses a for loop to read through the file line by line and, after finishing reading through the file, reports the **highest water level and the date and time** that was observed.



```python
# Answer Problem 1

# Open file name

import csv

cedar_key_water = 'CO-OPS_8727520_wl.csv'

# track the highest water level and its corresponding date and time
highest_level = float('-inf')
highest_level_time = ''

# Open the CSV file and read it
with open(cedar_key_water, mode='r') as file:
    csv_reader = csv.DictReader(file)
    
    for row in csv_reader:
        date = row['Date']
        time = row['Time (GMT)']
        observed_level = float(row['Preliminary (ft)'])
        
        # Checking the if the observed water level is the highest 
        if observed_level > highest_level:
            highest_level = observed_level
            highest_level_time = f'{date} {time}'

# Reporting the highest water level and the date and time it was observed
print(f"What is the highest water level and when did it occur?")
print("")
print(f'The highest water level observed was {highest_level} feet at {highest_level_time}.')
```

# Problem 2 (5 pts):

Either in a new script or modifying the above script, calculate the **lowest, highest and average** water level observed during the time period. As above, **print the date and time for the lowest and highest readings**. 



```python
# Answer Problem 2

import csv

cedar_key_water = 'CO-OPS_8727520_wl.csv'

# track the highest and lowest water level and its corresponding date and time
highest_level = float('-inf')
lowest_level = float('inf')
highest_level_time = ''
lowest_level_time = ''
total_level = 0
count = 0

# Open the CSV file and read it
with open(cedar_key_water, mode='r') as file:
    csv_reader = csv.DictReader(file)
    
    for row in csv_reader:
        date = row['Date']
        time = row['Time (GMT)']
        observed_level = float(row['Preliminary (ft)'])
        
        # Checking the if the observed water level is the highest 
        if observed_level > highest_level:
            highest_level = observed_level
            highest_level_time = f'{date} {time}'
            
        # Checking the if the observed water level is the lowest 
        if observed_level < lowest_level:
            lowest_level = observed_level
            lowest_level_time = f'{date} {time}'
            
   # total water level and number of observations
        total_level += observed_level
        count += 1

# average water level
average_level = total_level / count if count != 0 else 0

# Reporting the highest, lowest and average water level and the date and time it was observed
print(f"What is the highest water level and when did it occur?")
print("")
print(f'The highest water level observed was {highest_level} feet at {highest_level_time}.')
print("")
print(f"What is the lowest water level and when did it occur?")
print("")
print(f'The lowest water level observed was {lowest_level} feet at {lowest_level_time}.')
print("")
print(f"What is the average water level?")
print("")
print(f'The average water level observed was {average_level:.2f} feet.')
```

# Problem 3 (5 pts):

Write a script (or Jupyter Notebook) that calculates the **fastest rise in water level per 6-minute period between measurements** (for this assignment, assume that each line of the dataset is a 6-minute interval) and reports the **date and time** that was observed and the change in water level from the previous recording.



```python
#Answer Problem 3

import csv

cedar_key_water = 'CO-OPS_8727520_wl.csv'

# tracking the fastest rise in water levels and its corresponding date and time
fastest_rise = float("-inf")
fastest_rise_time = ""
previous_record = None

#Open the file and read it
with open (cedar_key_water, mode ="r") as file:
    csv_reader = csv.DictReader(file)
    
    for row in csv_reader:
        date = row["Date"]
        time = row["Time (GMT)"]
        observed_level = float(row["Preliminary (ft)"])
        
        #Rise in level water from previous records
        if previous_record is not None:
            rise = observed_level - previous_record
            
            #Checking the fastest rise in water level
            if rise > fastest_rise:
                fastest_rise = rise
                fastest_rise_time = f"{date} {time}"
                
        #Checking the current rise level
        previous_record = observed_level
                

# Reporting the fastest rise in water level and the date and time it was observed
print(f"What is the fastest rise in water level and when did it occur?")
print("")
print(f'The fastest rise in water level observed was {fastest_rise:.2f} feet at {fastest_rise_time}.')
```

* The water level increases more than 0.25 since the previous recording.
* The water level is over 5.0.
